```python
import numpy as np
import pandas as pd
from os import listdir
from os.path import isfile, join
import matplotlib.pyplot as plt
import math
from sklearn.model_selection import train_test_split
from sklearn.decomposition import TruncatedSVD
from sklearn.decomposition import PCA\
```

```python
preprocessed_df = pd.read_csv("./dataset/preprocessed_data.csv")
```

```python
preprocessed_df.head()
```

```python
preprocessed_df = preprocessed_df.drop(['Unnamed: 0'], axis=1)
```

```python
preprocessed_df.head()
```

```python
SVD = TruncatedSVD(n_components=15)
SVD.fit(preprocessed_df)
```

```python
preprocessed_df_svd_reduced = SVD.transform(preprocessed_df)
```

```python
preprocessed_df_svd_reduced.shape
```

```python
print(SVD.explained_variance_ratio_.sum())
```

```python
np.savetxt("preprocessed_data_SVD.csv", preprocessed_df_svd_reduced, delimiter=",")
```

```python
PCA = PCA(n_components=15).fit(preprocessed_df)
preprocessed_df_pca_reduced = PCA.transform(preprocessed_df)
```

```python
preprocessed_df_pca_reduced.shape
```

```python
print(PCA.explained_variance_ratio_.sum())
```

```python
np.savetxt("preprocessed_data_PCA.csv", preprocessed_df_pca_reduced, delimiter=",")
```
